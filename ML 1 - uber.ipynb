```python
import numpy as np
```

```python
def remove_outliers(df,col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3-q1
    lower = q1 - 1.5*iqr
    upper = q3 + 1.5*iqr
    df = df[(df[col] > lower) & (df[col] < upper)]
    return df
```

```python
def treat_outliers(df,col_list):
    for c in col_list:
        df = remove_outliers(df,c)
    return df
```

```python
df = treat_outliers(df,df.iloc[:, 0::])
```

```python
df.plot(kind='box', subplots=True, layout=(7,2), figsize=(15,20))
```

```python
import haversine as hs
```

```python
def dist(lat1,long1,lat2,long2):
    pt1 = (lat1,long1)
    pt2 = (lat2,long2)
    
    distance = hs.haversine(pt1,pt2)
    return distance
```

```python
df['distance'] = df.apply(lambda row: dist(row['pickup_latitude'], row['pickup_longitude'], row['dropoff_latitude'], row['dropoff_longitude']), axis=1)
```

```python
df = df[df['fare_amount']>0]
```

```python
df.drop(columns=['pickup_longitude','pickup_latitude','dropoff_latitude','dropoff_longitude'],inplace=True)
```

```python
x = df.drop(['fare_amount'],axis=1)
y = df['fare_amount']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred_lr = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_lr = r2_score(y_test,y_pred_lr)
mse_lr = mean_squared_error(y_test,y_pred_lr)
print(r2_lr)
print(mse_lr)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=101)
rf.fit(x_train,y_train)
```

```python
y_pred_rf = rf.predict(x_test)
```
